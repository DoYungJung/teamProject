```python
#라이브러리 선언
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import bs4
import time
import pandas as pd

#드라이버 위치 설정
driver_loc = "../externlib/chromedriver/chromedriver.exe"

#드라이버 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("window-size=1920x1080")

#웹 드라이버 정의
driver =webdriver.Chrome(driver_loc , options=options)
# driver.implicitly_wait(3)

##Tripadvisor 사이트에서 서울 음식점 카테고리 접속
baseUrl = "https://www.tripadvisor.co.kr"
Seoul = "/Restaurants-g294197-Seoul.html"
driver.get(baseUrl+Seoul)
time.sleep(2)

#강남구 클릭
gangnamgu = '//*[@id="component_48"]/div/div[10]/div[2]/div[1]/div/label'
driver.find_element_by_xpath(gangnamgu).click()
time.sleep(2)

#신사동 클릭
sinsadong = '//*[@id="component_48"]/div/div[10]/div[2]/div[4]/div/label/div/span/span'
driver.find_element_by_xpath(sinsadong).click()
time.sleep(2)

#강남구 클릭 해제
gangnam = '//*[@id="component_48"]/div/div[10]/div[2]/div[1]/div/label/div/span/span'
driver.find_element_by_xpath(gangnam).click()
time.sleep(2)


```

```python
rowList = []
```

```python
j=19
while(1==1):
 
    ##현재 페이지 소스 가져오기
    html = driver.page_source
    time.sleep(2)
    ## BeautifulSoup로 페이지 소스 파싱
    bs = bs4.BeautifulSoup(html, "html.parser")
    time.sleep(2)
    #div태그 중 필요한 부분 추출
    divTag = bs.find("div" , attrs={"class":"deQwQ"})
    time.sleep(2)
    divLen = len(divTag.findAll(attrs={"class":"cauvp Gi o"}))
    
    for i in range(0, divLen):
        list =[]
        try:
            hrefAttrs = divTag.findAll(attrs={"class":"cauvp Gi o"})[i].find(attrs={"class":"bHGqj Cj b"}).attrs["href"]
            time.sleep(2)
            targetUrl = baseUrl+hrefAttrs
            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[1])   
            driver.get(targetUrl)
            time.sleep(1)

            newHtml = driver.page_source
            time.sleep(1)

            newBs = bs4.BeautifulSoup(newHtml, "html.parser")
            time.sleep(2)
        
        
            ##식당명
            name = newBs.find(attrs={"data-test-target":"top-info-header"}).text

            ##평점 추출
            score = newBs.find(attrs={"class":"fdsdx"}).text[0:3]

            ##리뷰 개수

            reviewCnt = newBs.find(attrs={"class":"eBTWs"}).text.replace("건의 리뷰", "")

            ##요리 종류

            category=""
            if(newBs.find(attrs={"class":"cfvAV"}).text.split(",")[0][0] == '₩'):
                category = newBs.findAll(attrs={"class":"cfvAV"})[1].text.split(",")[0]

            else:
                category = newBs.find(attrs={"class":"cfvAV"}).text.split(",")[0]

            ## 주소
            address = newBs.find(attrs={"href":"#MAPVIEW"}).text


            ##운영시간
            ##시작 시간의 오전 오후 여부
            ##startTime = newBs.find(attrs={"class":"cSjwK"}).find("span").findAll("span")[1].find("span")

            ##시작시간
            # startTimeAmPm = newBs.find(attrs={"class":"cSjwK"}).find("span").findAll("span")[1].find("span").find("span").text

            ##임시 방편
            #start = startTime + " " +startTimeAmPm
            # start =""


            ##끝나는 시간 오전 오후 여부
            # endTime = newBs.find(attrs={"class":"cSjwK"}).findAll("span")[9].text[-2:]

            ##끝나는 시간

            # endTimeAmPm = newBs.find(attrs={"class":"cSjwK"}).findAll("span")[8].text

            #end = endTime + " " +endTimeAmPm

            # end =""


            list = [name, score, reviewCnt, category, address]
            time.sleep(1)
            rowList.append(list)
            time.sleep(1)
        except:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            time.sleep(2)
            continue
            
        list =[]
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(2)
        
    if(j==0):
        daum = '//*[@id="EATERY_LIST_CONTENTS"]/div[2]/div/a'
        driver.find_element_by_xpath(daum).click()
        time.sleep(2)
        j=j+1
    else:
        try:
            daum='//*[@id="EATERY_LIST_CONTENTS"]/div[2]/div/a[2]'
            driver.find_element_by_xpath(daum).click()
            time.sleep(2)
            j=j+1
        except:
            break

    
```

```python
pd.DataFrame(rowList, columns=["이름", "평점", "리뷰 수", "카테고리", "주소"])
```

```python
result = pd.DataFrame(rowList, columns=["이름", "평점", "리뷰 수", "카테고리", "주소"])
```
