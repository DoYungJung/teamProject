### 데이터 불러오기

```python
### CSV 파일을 읽어 DataFrame 변수에 저장하기
### 1페이지_이름 순 정렬
ageData1 = pd.read_csv("../we/gundae1_age_gender_final.csv", encoding="ms949")

### 2페이지_이름 순 정렬
ageData2 = pd.read_csv("../we/gundae2_age_gender_final.csv", encoding="ms949")

### 3페이지_이름 순 정렬
ageData3 = pd.read_csv("../we/gundae3_age_gender_final.csv", encoding="ms949")

### 4페이지_이름 순 정렬
ageData4 = pd.read_csv("../we/gundae4_age_gender_final.csv", encoding="ms949")

### 5페이지_이름 순 정렬
ageData5 = pd.read_csv("../we/gundae5_age_gender_final.csv", encoding="ms949")

### 6페이지_이름 순 정렬
ageData6 = pd.read_csv("../we/gundae6_age_gender_final.csv", encoding="ms949")
```

### 1. 데이터 병합

```python
merge12 = pd.merge(ageData1, ageData2, how="outer", on=None)
merge34 = pd.merge(ageData3, ageData4, how="outer", on=None)
merge56 = pd.merge(ageData5, ageData6, how="outer", on=None)
merge1234 = pd.merge(merge12,merge34, how="outer", on=None)

mergeData1 = pd.merge(merge1234,merge56, how="outer", on=None)
```

```python
# 중복 제거
mergeData1.drop_duplicates()
```

```python
mergeData1.shape
```

```python
mergeData1.info()
```

```python
### 데이터 전처리
mergeData1.tenCen = mergeData1.tenCen.str[0:-5]
mergeData1.twoCen = mergeData1.twoCen.str[0:-5]
mergeData1.threeCen = mergeData1.threeCen.str[0:-5]
mergeData1.fourCen = mergeData1.fourCen.str[0:-5]
mergeData1.fiveCen = mergeData1.fiveCen.str[0:-5]
mergeData1.sixCen = mergeData1.sixCen.str[0:-5]
```

```python
mergeData1.womanRatio = mergeData1.womanRatio.str[0:-1]
mergeData1.womanRatio = mergeData1.womanRatio.str[-2:]
mergeData1.manRatio = mergeData1.manRatio.str[0:-1]
mergeData1.manRatio = mergeData1.manRatio.str[-2:]
```

```python
mergeData1
```

```python
mergeData1.info()
```

```python
mergeData1["tenCen"] = mergeData1.tenCen.astype(float)
mergeData1["twoCen"] = mergeData1.twoCen.astype(float)
mergeData1["threeCen"] = mergeData1.threeCen.astype(float)
mergeData1["fourCen"] = mergeData1.fourCen.astype(float)
mergeData1["fiveCen"] = mergeData1.fiveCen.astype(float)
mergeData1["sixCen"] = mergeData1.sixCen.astype(float)

mergeData1["womanRatio"] = mergeData1.womanRatio.astype(int)
mergeData1["manRatio"] = mergeData1.manRatio.astype(int)
```

```python
mergeData1
```

```python
### 타입 통합
mergeData1 =  mergeData1.astype( { "tenCen" : float, "twoCen" : float, "threeCen" : float, "fourCen" : float, "fiveCen" : float, "sixCen" : float, "womanRatio" : int, "manRatio" : int })
```

```python
mergeData1 = mergeData1.sort_values(by=["name"], ignore_index=True)
```

```python
mergeData1 = mergeData1.drop(columns=["Unnamed: 0"])
```
