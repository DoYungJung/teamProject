```python
import pandas as pd
```

```python
regionData = pd.read_csv("../dataset/analysis_list/region_code.csv")
loadShopData = pd.read_csv("../dataset/analysis_list/seoul_load_shop_list.csv")
peopleData = pd.read_csv("../dataset/analysis_list/seoul_move_people_list.csv")
restaurantData = pd.read_csv("../dataset/analysis_list/sinsa_restaurant_list.csv")
```

```python
loadShopData
```

```python
regionData
```

```python
### 상권지역 상세 법정동 조인
detailLoadData = pd.merge(left=loadShopData,
                          right=regionData,
                          left_on=["ADSTRD_CD"],
                          right_on=["ADSTRD_CD"], how="left")
```

```python
detailLoadData
```

```python
### 서울 강남구 신사동 내 상권 정보 추출
sinsaLoadData = detailLoadData.loc[(detailLoadData["REGION_SI"] == "서울특별시") & (detailLoadData["REGION_GU"] == "강남구") & (detailLoadData["REGION_DONG"] == "신사동")]
```

```python
sinsaLoadData
```

```python
loadRestauranData = pd.merge(left=restaurantData,
                            right=sinsaLoadData,
                            left_on=["ADSTRD_CD"],
                            right_on=["ADSTRD_CD"], how="right")
```

```python
### 상권 구역별 음식점 리스트 조인
### 나중에 여러분들이 수집한 음식점 리스트 주소를 여기에 조인
hanNamIc = loadRestauranData.loc[(loadRestauranData["TRDAR_CD_NM"] == "한남IC")]

sinsaPark = loadRestauranData.loc[(loadRestauranData["TRDAR_CD_NM"] == "신사은행나무공원")]

garosugill = loadRestauranData.loc[(loadRestauranData["TRDAR_CD_NM"] == "가로수길")]

apgujungSt = loadRestauranData.loc[(loadRestauranData["TRDAR_CD_NM"] == "압구정역")]

gangnamMk = loadRestauranData.loc[(loadRestauranData["TRDAR_CD_NM"] == "강남시장(강남구 강남시장)")]
```

```python
peopleLoadData = pd.merge(left=sinsaLoadData,
                            right=peopleData,
                            left_on=["TRDAR_CD"],
                            right_on=["TRDAR_CD"], how="left").drop_duplicates(["TRDAR_CD"])
```

```python
### 상권 구역별 연령, 성별 유동인구표
### 머신러닝때 사용자가 연령대를 입력하면 5개 지역 중 연령대에 맞게 상권 구역과 음식점 추천 로직으로 만들면 좋을듯... 할 수 있다면...

hanNamPeople = peopleLoadData.loc[(peopleLoadData["TRDAR_CD"] == 2110949)]

sinsaParkPeople = peopleLoadData.loc[(peopleLoadData["TRDAR_CD"] == 2110951)]

garosugillPeople = peopleLoadData.loc[(peopleLoadData["TRDAR_CD"] == 2120186)]

apgujungStPeople = peopleLoadData.loc[(peopleLoadData["TRDAR_CD"] == 2120188)]

gangnamMkPeople = peopleLoadData.loc[(peopleLoadData["TRDAR_CD"] == 2130302)]
```

```python
hanNamPeople
```

```python
### 서울 강남구 신사동 내 상권 정보
sinsaLoadData.to_csv('../dataset/analysis_list/sinsa_load_shop_list.csv', index=False, encoding="utf-8-sig")

### 상권 구역별 음식점 리스트 
hanNamIc.to_csv('../dataset/analysis_list/hanNamIc_load_restaurant_list.csv', index=False, encoding="utf-8-sig")
sinsaPark.to_csv('../dataset/analysis_list/sinsaPark_load_restaurant_list.csv', index=False, encoding="utf-8-sig")
garosugill.to_csv('../dataset/analysis_list/garosugill_load_restaurant_list.csv', index=False, encoding="utf-8-sig")
apgujungSt.to_csv('../dataset/analysis_list/apgujungSt_load_restaurant_list.csv', index=False, encoding="utf-8-sig")
gangnamMk.to_csv('../dataset/analysis_list/gangnamMk_load_restaurant_list.csv', index=False, encoding="utf-8-sig")
loadRestauranData.to_csv('../dataset/analysis_list/sinsa_load_restaurant_list.csv', index=False, encoding="utf-8-sig")

### 상권 구역별 연령, 성별 유동인구표 
hanNamPeople.to_csv('../dataset/analysis_list/hanNamIc_people_list.csv', index=False, encoding="utf-8-sig")
sinsaParkPeople.to_csv('../dataset/analysis_list/sinsaPark_people_list.csv', index=False, encoding="utf-8-sig")
garosugillPeople.to_csv('../dataset/analysis_list/garosugill_people_list.csv', index=False, encoding="utf-8-sig")
apgujungStPeople.to_csv('../dataset/analysis_list/apgujungSt_people_list.csv', index=False, encoding="utf-8-sig")
gangnamMkPeople.to_csv('../dataset/analysis_list/gangnamMk_people_list.csv', index=False, encoding="utf-8-sig")
peopleLoadData.to_csv('../dataset/analysis_list/sinsa__people_list.csv', index=False, encoding="utf-8-sig")
```
