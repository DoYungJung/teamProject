```python
import pandas as pd
```

```python
loadList = pd.read_csv("../dataset/서울시 우리마을가게 상권분석서비스(상권영역).csv", encoding="ms949")
```

```python
loadList
```

```python
### 행정동 코드 자리 수 불일치로 행정동 코드 뒤에 '00'을 추가하여 자릿 수 보정
loadList["행정동_코드_보정치"] = loadList["행정동_코드"].apply(str) + '00'

loadList["행정동_코드_보정치"][0]

loadList.astype({"행정동_코드_보정치" : 'int64'}).dtypes
```

```python
### 리스트 컬럼조회
loadColList = loadList.columns

loadColList
```

```python
### 컬럼 리스트 DF화
columnsList = pd.DataFrame(columns=['CATEGORY'])

### 반복조회 후 저장
for i in range (0, len(loadColList)):
    columnsList = columnsList.append({'CATEGORY' :loadColList[i]}, ignore_index=True)
```

```python
### 서울시 상권 리스트
seoulSLoadShopList = pd.DataFrame(columns=['STDR_YM_CD', 'TRDAR_SE_CD', 'TRDAR_SE_CD_NM', 'TRDAR_CD', 'TRDAR_CD_NM', 'SIGNGU_CD', 'ADSTRD_CD'])

### 반복조회 후 저장
for i in range (0, len(loadList)):
    seoulSLoadShopList = seoulSLoadShopList.append({'STDR_YM_CD' :loadList["기준_년월_코드"].values[i],\
                                                    'TRDAR_SE_CD' : loadList["상권_구분_코드"].values[i],\
                                                    'TRDAR_SE_CD_NM' : loadList["상권_구분_코드_명"].values[i],\
                                                    'TRDAR_CD' : loadList["상권_코드"].values[i],\
                                                    'TRDAR_CD_NM' : loadList["상권_코드_명"].values[i],\
                                                    'SIGNGU_CD' : loadList["시군구_코드"].values[i],\
                                                    'ADSTRD_CD' : loadList["행정동_코드_보정치"].values[i]}, ignore_index=True)

### 데이터 저장 확인
seoulSLoadShopList
```

```python
seoulSLoadShopList.to_csv('../dataset/analysis_list/seoul_load_shop_list.csv', index=False, encoding="utf-8-sig")
```
