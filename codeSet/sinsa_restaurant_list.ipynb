```python
### 사용 라이브러리
import pandas as pd
```

```python
### csv데이터 불러오기
shopList = pd.read_csv("../dataset/소상공인시장진흥공단_상가(상권)정보_서울_202112.csv")

shopList
```

```python
### 리스트 컬럼조회
shopColList = shopList.columns

shopColList
```

```python
### 컬럼 리스트 DF화
columnsList = pd.DataFrame(columns=['CATEGORY'])

### 반복조회 후 저장
for i in range (0, len(shopColList)):
    columnsList = columnsList.append({'CATEGORY' :shopColList[i]}, ignore_index=True)

### 신사동 음식점 리스트 저장
restaurantList = shopList.loc[ (shopList["시군구명"]=="강남구") & (shopList["상권업종대분류명"]=="음식") & (shopList["법정동명"]=="신사동") ]
```

```python
### 신사동 음식점 상세리스트 저장 
sinsaRestaurantList = pd.DataFrame(columns=['SHOP_CODE', 'SHOP_NAME', 'SHOP_CATEGORY', 'ADSTRD_CD', 'REGION_ID', 'SHOP_ADR'])

### 반복조회 후 저장
for i in range (0, len(restaurantList)):
    sinsaRestaurantList = sinsaRestaurantList.append({'SHOP_CODE' :restaurantList["상가업소번호"].values[i],\
                                                    'SHOP_NAME' : restaurantList["상호명"].values[i],\
                                                    'SHOP_CATEGORY' : restaurantList["상권업종소분류명"].values[i],\
                                                    'ADSTRD_CD' : restaurantList["행정동코드"].values[i],\
                                                    'REGION_ID' : restaurantList["시도명"].values[i] + " " + restaurantList["시군구명"].values[i] + " " + restaurantList["법정동명"].values[i],\
                                                    'SHOP_ADR' : restaurantList["도로명주소"].values[i]}, ignore_index=True)

### 데이터 저장 확인
sinsaRestaurantList
```

```python
sinsaRestaurantList.to_csv('../dataset/analysis_list/sinsa_restaurant_list.csv', index=False, encoding="utf-8-sig")
```
