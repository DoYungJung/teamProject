### 데이터 불러오기

```python
### CSV 파일을 읽어 DataFrame 변수에 저장하기
### 1페이지_이름 순 정렬
ageData1 = pd.read_csv("../we/gundae6_age_gender_final.csv", encoding="ms949")
```

### 1. 데이터 전처리

```python
# 중복 제거
ageData1.drop_duplicates()
```

```python
ageData1.shape
```

```python
ageData1.info()
```

```python
### 데이터 전처리
ageData1.tenCen = ageData1.tenCen.str[0:-5]
ageData1.twoCen = ageData1.twoCen.str[0:-5]
ageData1.threeCen = ageData1.threeCen.str[0:-5]
ageData1.fourCen = ageData1.fourCen.str[0:-5]
ageData1.fiveCen = ageData1.fiveCen.str[0:-5]
ageData1.sixCen = ageData1.sixCen.str[0:-5]
```

```python
ageData1.womanRatio = ageData1.womanRatio.str[0:-1]
ageData1.womanRatio = ageData1.womanRatio.str[-2:]
ageData1.manRatio = ageData1.manRatio.str[0:-1]
ageData1.manRatio = ageData1.manRatio.str[-2:]
```

```python
ageData1
```

```python
ageData1.info()
```

```python
ageData1["tenCen"] = ageData1.tenCen.astype(float)
ageData1["twoCen"] = ageData1.twoCen.astype(float)
ageData1["threeCen"] = ageData1.threeCen.astype(float)
ageData1["fourCen"] = ageData1.fourCen.astype(float)
ageData1["fiveCen"] = ageData1.fiveCen.astype(float)
ageData1["sixCen"] = ageData1.sixCen.astype(float)

ageData1["womanRatio"] = ageData1.womanRatio.astype(int)
ageData1["manRatio"] = ageData1.manRatio.astype(int)
```

```python
ageData1
```

```python
### 타입 통합
ageData1 =  ageData1.astype( { "tenCen" : float, "twoCen" : float, "threeCen" : float, "fourCen" : float, "fiveCen" : float, "sixCen" : float, "womanRatio" : int, "manRatio" : int })
```

```python
ageData1 = ageData1.sort_values(by=["name"], ignore_index=True)
```

```python
ageData1 = ageData1.drop(columns=["Unnamed: 0"])
```
