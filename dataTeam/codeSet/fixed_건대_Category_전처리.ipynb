### 1. 데이터 병합

```python
# 중복 제거
ageData1.drop_duplicates()
```

```python
ageData1.shape
```

```python
ageData1.info()
```

```python
### 데이터 전처리
ageData1["score"] = ageData1.score.str.split('/').str[0] # 별점 점수
ageData1["visitCount"] = ageData1.visitCount.str.split('_').str[1] # 방문자 리뷰 수
ageData1["blogCount"] = ageData1.blogCount.str.split('_').str[1] # 블로그 리뷰 수
```

```python
### 리뷰 수 콤마 제거 (int형으로 변환하기위함)
ageData1["visitCount"] = ageData1.visitCount.str.replace(',', '')
ageData1["blogCount"] = ageData1.blogCount.str.replace(',', '')
```

```python
ageData1.visitCount.loc[ageData1.visitCount.str.contains("주문자리뷰")] = 147
```

```python
### 타입 통합
ageData1 =  ageData1.astype( { "score" : float, "visitCount" : int, "blogCount" : int })
```

```python
ageData1 = ageData1.sort_values(by=["name"], ignore_index=True)
```
