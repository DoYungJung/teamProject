```python
select_root = driver.find_elements_by_class_name('link')
dataFramesList = []
for i in range(len(select_root)):
    needUrl = select_root[i].get_attribute('href')
    time.sleep(2)
    driver.get(needUrl)
    time.sleep(2)

    restrauntsName = driver.find_elements_by_class_name('restaurant-details__heading--title')
    time.sleep(2)
    locationOfRest=driver.find_elements_by_class_name('restaurant-details__heading--list')
    
    restrauntsNameList =[]
    locationOfRestList=[]
    priceList =[]
    foodList = []
    finallocationOfRestList=[]

    for i in range(len(restrauntsName)):
        if(restrauntsName[i].text != ""):
            restrauntsNameList.append(restrauntsName[i].text)
        if(locationOfRest[i].text != ""):
            refineLocationOfRest = locationOfRest[i].text.split("\n")
            finallocationOfRestList.append(refineLocationOfRest[0])
            secondlocationOfRestList = refineLocationOfRest[1].split("•")
            priceList.append(secondlocationOfRestList[0])
            foodList.append(secondlocationOfRestList[1])

    restDescription=driver.find_element_by_class_name('js-show-description-text')
    descriptionList = [restDescription.text]
    time.sleep(2)
    michelinGrade=driver.find_element_by_class_name('restaurant-details__classification--list')
    michelinGradeList = michelinGrade.text.split("\n")
    finalMichelinGradeList=[]
    time.sleep(2)
    for i in range(len(michelinGradeList)):
        if(len(str(michelinGradeList[i])) != 1):
            finalMichelinGradeList.append(michelinGradeList[i])

    serviceAndfacilityList = []
    serviceAndfacility=driver.find_elements_by_class_name('restaurant-details__services--content')
    time.sleep(2)
    for i in range(len(serviceAndfacility)):
        tmp=""
        tmp=serviceAndfacility[i].text.split("\n")
        if(len(tmp) != 1):
            tmp=tmp[1]
        else:
            tmp = tmp[0]
        print(tmp)
        serviceAndfacilityList.append(tmp)

    tellphoneNum =driver.find_elements_by_class_name('flex-fill')
    tellphoneNumber = tellphoneNum[1].text
    tellphoneNumberList = [tellphoneNumber]

    # 데이터 프레임으로 변환
    time.sleep(1)
    restName = restrauntsNameList[0]
    price = priceList[0]
    foodtype = foodList[0]
    location = finallocationOfRestList[0]
    Grade = finalMichelinGradeList[0]
    description = descriptionList[0]
    specific = ""
    for i in range(len(serviceAndfacilityList)):
        specific += serviceAndfacilityList[i] + " "

    tellNum = tellphoneNumberList[0]
    result = [restName,price,foodtype,location,Grade,description,specific,tellNum]
    columns = ["가게이름", "가격", "카테고리", "위치","등급","특징","부가서비스","전화번호"]
    dataFrame = pd.DataFrame(data = [result], columns=columns)
    dataFramesList.append(dataFrame)
    driver.back()

michelin3Frame = pd.concat(dataFramesList, ignore_index = True)
```
